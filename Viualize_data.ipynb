```python
import pandas as pd
from config import *
import matplotlib.pyplot as plt

# name_appliance = 'dishwasher'
name_appliance = 'kettle'

sequence_length = APPLIANCE_CONFIG[name_appliance]['window_width'] // SAMPLE_WIDTH
```

```python
csv_path = 'prepocessed_data/dataset_' + name_appliance + '.csv'
df = pd.read_csv(csv_path, sep='\s+')
```

```python
applience_sequence_head = ['appliance_power_' + str(i) for i in range(sequence_length)]
aggregate_sequence_head = ['aggregate_power_' + str(i) for i in range(sequence_length)]
```

```python
for i in range(20):
    appliance_power = df.loc[i, applience_sequence_head].values
    aggregate_power = df.loc[i, aggregate_sequence_head].values
    
    plt.plot(appliance_power, label="appliance")
    plt.plot(aggregate_power, label="aggregate")
    plt.xlabel('Time (S)')
    plt.ylabel('Power (KW/H)')
    plt.legend()
    plt.show()
```
